# deteksi outlier (local outlier factor)

outlier merujuk pada titik data yang secara nyata berbeda dari mayoritas data lainnya dalam kumpulan data. keberadaan outlier seringkali disebabkan oleh kesalahan pengukuran, fenomena anomali, atau variasi yang sah dalam data yang diamati. dalam analisis data, keberadaan outlier bisa memiliki dampak yang penting pada hasil analisis, termasuk mengubah statistik deskriptif, estimasi parameter, dan kinerja model prediktif. salah satu pendekatan yang digunakan untuk mengidentifikasi outlier dalam dataset multidimensional adalah metode yang dikenal sebagai local outlier factor (lof). lof bekerja dengan mengevaluasi tingkat anomali lokal untuk setiap titik data dengan membandingkan kepadatan titik data di sekitarnya dengan kepadatan tetangganya. jika tingkat kepadatan titik data tersebut lebih rendah secara relatif dibandingkan dengan tetangganya, kemungkinan besar titik data tersebut dapat dianggap sebagai outlier.

## konsep-konsep outlier

### k-distance dan k-neighbors


k-distance adalah jarak antara suatu titik data dengan titik data terdekat ke-k, dengan nilai k telah ditentukan sebelumnya. fokus dari perhitungan k-distance adalah untuk menilai kerapatan lokal sebuah titik data dengan mengukur jaraknya terhadap tetangga-tetangganya. semakin besar nilai k-distance, semakin rendah tingkat kerapatan lokal titik data tersebut.
sekelompok titik data yang terdekat dengan sebuah titik data tertentu disebut sebagai k-neighbors. dalam konteks lof, tetangga k digunakan untuk menghitung kepadatan lokal (local reachability density/lrd) dan faktor outlier lokal (local outlier factor/lof). jumlah k-neighbors yang digunakan akan berpengaruh pada hasil deteksi outlier. semakin besar nilai k, semakin banyak tetangga yang akan dipertimbangkan dalam perhitungan lrd dan lof. meskipun ini dapat menghasilkan estimasi yang lebih stabil, namun juga memerlukan waktu komputasi yang lebih lama.

<img src="image-20240321-042402.png" width="" align="" />

contoh tersebut jika k=3 maka tetangga terdekatnya 2 class B dan satu kelas A, dan jika k=6 maka tetangga terdekatnya yaitu 4 class A dan 3 class B

### reachability distance (RD)

merupakan jarak k maksimum dari titik tetangga dan jarak antara dua titik. RD = max(k-distance(B),distance(A,B))

### local reachability density (LRD)

rumus umumnya : 

<img src="image-20240321-042802.png" width="" align="" />

x adalah titik data yang di evaluasi, y adalah salah satu dari k-neighbors terdekat dari x, dan RD(x,y) adalah reachability distance antara x dan y tinggi nilai LRD sebuah titik data, semakin dekat titik data tersebut dengan tetangga-tetangganya.

### local outlier factor (LOF)

<img src="image-20240321-043310.png" width="" align="" />


lrd dari setiap titik digunakan untuk membandingkannya dengan rata-rata lrd dari tetangga k-nya. lof merupakan perbandingan antara rata-rata lrd dari tetangga k sebuah titik a dengan lrd dari titik a itu sendiri.
jika suatu titik bukan outlier (inlier), rata-rata lrd dari tetangga k-nya akan mirip dengan lrd dari titik tersebut (karena kerapatan titik dan tetangganya hampir sama). dalam situasi ini, lof akan mendekati nilai 1. sebaliknya, jika suatu titik adalah outlier, lrd dari titik tersebut akan lebih rendah dari rata-rata lrd dari tetangga k-nya. dengan demikian, nilai lof akan tinggi.
meskipun lof > 1 sering dianggap sebagai tanda adanya outlier, namun ini tidak selalu benar. misalnya, jika kita mengetahui bahwa hanya ada satu outlier dalam data, maka kita bisa mengambil nilai lof maksimum dari semua titik dan mempertimbangkan titik yang sesuai dengan nilai lof maksimum sebagai outlier.

## menghitung menggunakan python (scikit learn)

setelah menentukan outlier manual menggunakan excel, selanjutnya akan menguji hasil tersebut menggunakan scikit learn

```
from sklearn.neighbors import LocalOutlierFactor
import numpy as np

X = [
    (2, 15, 1),
    (1.5, 16, 0.5),
    (1, 15, 2),
    (3, 17, 3),
    (2, 15, 1.5),
    (20, 50, 10),
    (4, 16, 2),
    (2.5, 14, 3),
    (4, 16, 1),
    (3.7, 17, 4),
]

lof = LocalOutlierFactor(n_neighbors=3, contamination=0.1)

y_pred = lof.fit_predict(X)

print("Predicted table:", y_pred)
print("Negative LOF scores:", -lof.negative_outlier_factor_)  # Ubah tanda negatif untuk mendapatkan nilai LOF positif
```

```

!pip install mysql-connector-python
from sklearn.neighbors import LocalOutlierFactor
import mysql.connector
import matplotlib.pyplot as plt

# Membuat koneksi ke database MySQL
conn = mysql.connector.connect(
    host="datamining-tasya22022.a.aivencloud.com",
    user="avnadmin",
    password="AVNS_bsya-rWQfyVA9CtdGxk",
    port = "24072",
    database="db_sql"
)

# Membaca data dari database menggunakan Pandas
query = "SELECT SepalLengthCm, SepalWidthCm FROM outlier"
cursor = conn.cursor()
cursor.execute(query)
rows = cursor.fetchall()
x = [[row[0], row[1]] for row in rows]

# Menutup koneksi ke database
conn.close()

# Membuat model LOF
lof = LocalOutlierFactor(n_neighbors=13, contamination=0.03)

# Melakukan prediksi outlier
y_pred = lof.fit_predict(x)

# Menampilkan data dengan outlier yang diidentifikasi
outlier_indices = [i for i, label in enumerate(y_pred) if label == -1]
values = [x[i] for i in outlier_indices]

plt.scatter([point[0] for point in x], [point[1] for point in x])
plt.scatter([point[0] for point in values], [point[1] for point in values], color='r')
plt.show()
```
